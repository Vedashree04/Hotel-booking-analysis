## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# create a copy of current dataset.
df= hotel_df.copy()
````

````python
# dropping duplicate values
hotel_df.drop_duplicates(inplace = True)
df.shape
````

````python
# handling missing values
hotel_df.isnull().sum().sort_values(ascending = False)[:6]
````

````python
hotel_df[['company','agent']] = hotel_df[['company','agent']].fillna(0)
hotel_df['children'].unique()
hotel_df['children'].fillna(hotel_df['children'].mean(), inplace = True)     # We will replace the null values under this column with mean value of children.
hotel_df['country'].fillna('others', inplace = True)
# the detail explaination will be given below.
````

````python
# checking if all null values are removed
hotel_df.isnull().sum().sort_values(ascending = False) [:6]
````

````python
hotel_df.dtypes
````

````python
# In country we can not set the country values as 0 becuase it contains country codes representing different countries.
#So instead of replacing it with zero we have replaced it with the mode value in country column.
#Mode is nothing but just the most repeating value .

hotel_df.country.mode().to_string()
````

````python
hotel_df['country'].fillna(hotel_df.country.mode().to_string(), inplace= True)  # we use "inplace = True" to make permanant change
````

````python
# There are some rows with total number of adults, children or babies equal to zero. So we will remove such rows.
hotel_df[hotel_df['adults']+ hotel_df['babies']+ hotel_df['children'] == 0].shape
````

````python
hotel_df.drop(hotel_df[hotel_df['adults']+ hotel_df['babies']+ hotel_df['children'] == 0].index, inplace =True)
````

````python
# Converting columns to appropriate datatypes.
# Converting datatype of columns 'children', 'company' and 'agent' from float to int.
hotel_df[['children', 'company', 'agent']] = hotel_df[['children', 'company', 'agent']].astype(int)
````

````python
# changing datatype of column 'reservation_status_date' to data_type.
hotel_df['reservation_status_date'] = pd.to_datetime(hotel_df['reservation_status_date'], format = '%Y-%m-%d')
````

````python
# Adding important columns
# Adding total staying days hotels
hotel_df['total_stay'] = hotel_df['stays_in_weekend_nights'] + hotel_df['stays_in_week_nights']
````

````python
# adding total people of numbers as column , total people = num of adults + children + babeis
hotel_df['total_pople'] = hotel_df['adults'] + hotel_df['children'] + hotel_df['babies'] # We are adding this column so that we can analyse the stay length at hotels.
````

````python
# num_df1 = df1[['lead_time','previous_cancellations','previous_bookings_not_canceled','booking_changes','days_in_waiting_list','adr','required_car_parking_spaces','total_of_special_requests','total_stay','total_people']]
#Changing the data type of total people column from float to int
hotel_df['total_pople'] = hotel_df['total_pople'].astype(int)
````

````python
# findings on how longer people stays at the hotels

hotel_df[hotel_df['is_canceled'] == 0]['total_stay'].mean()
````

````python
filter = hotel_df['is_canceled'] == 0
data = hotel_df[filter]
data.head()
````
